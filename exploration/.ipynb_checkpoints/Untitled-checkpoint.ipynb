```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
pair = "EUR_USD"
granularity = "H4"
df = pd.read_pickle(f"../data/{pair}_{granularity}.pkl")
```

```python
df.tail()
```

```python
# take just 100 items from the api
df_plot = df.iloc[:100]
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x = df_plot.time,
    open= df_plot.mid_o,
    high= df_plot.mid_h,
    low= df_plot.mid_l,
    close= df_plot.mid_c
))
```
